```python
import matplotlib.pyplot as plt

wows = [12.22, 33.38, 47.10]
nqa = [46.92, 58.69, 63.70]
dstc9 = [7.47, 12.18, 15.86]
wowp = [68.68, 79.22, 82.73]

x = [1, 3, 5]

plt.plot(x, nqa, label='QA', marker='^', markersize=8)
plt.plot(x, dstc9, label='TOD', marker='s', markersize=8)
plt.plot(x, wows, label='KGD (Sentence)', marker='o', markersize=8)
plt.plot(x, wowp, label='KGD (Paragraph)', marker='D', markersize=8)

plt.xticks(x, ['1', '3', '5'], fontsize=12)
plt.yticks(range(0, 110, 10), fontsize=11)
plt.ylim(0, 107)
plt.grid(axis='y')
plt.xlabel('Number of documents retrieved (K)', fontsize=16)
plt.ylabel('Recall@K (%)', fontsize=16)
plt.title('Recall@K for different Dialogue Settings', fontsize=16, y=1.02)

plt.legend(loc='upper left', fontsize=11, facecolor='white', framealpha=1)

plt.savefig('recall_at_k.pdf', bbox_inches='tight')
```
